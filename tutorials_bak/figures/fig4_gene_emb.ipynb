```python
import matplotlib.pyplot as plt
import pandas as pd
import os
if not os.path.exists('./fig4'):
    os.makedirs('./fig4',  exist_ok=True)

df = pd.read_excel('../plots/figs_data.xlsx', sheet_name='fig4-gene_resource')

fig = plt.figure(figsize=(8, 3))
plt.subplots_adjust(wspace=0.4, hspace=0)
axes_1 = plt.subplot(121)
axes_1.bar('Name', 'GPU', data=df, color='#f29600')
plt.xticks(rotation=30)
axes_1.set_ylabel('GPU Memory (G)')

axes_2 = plt.subplot(122)
axes_2.bar('Name', 'times', data=df, color='#4974a4')
plt.xticks(rotation=30)
axes_2.set_ylabel('Running times (s)')


# df.plot.bar(x='model', y='GPU', color="#a4cde1", )
# plt.savefig('./fig2b.pdf', bbox_inches='tight')

plt.savefig('./fig4/1.gene_emb_resource.pdf', bbox_inches='tight')
```
